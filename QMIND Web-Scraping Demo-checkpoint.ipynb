```python
# Need to start by first importing the libraries that we will need to do the scraping
# For the purposes of this simple demo, will need beautifulsoup4 and requests library
from bs4 import BeautifulSoup
import requests
```

```python
# Now we will start by downloading the HTML data from the target website
# This will be done by using the get method from the Python Requests library
page = requests.get("https://www.yelp.ca/biz/stages-nightclub-kingston?osq=Stages")
```

```python
# Now we want to be able to read this data out
# Will now assign a new variable as the page content using BeautifulSoup html parser
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify)
```

```python
# Now we wish to find out where all of our interesting data is, since I want to get reviews,
# I know to use find_all('p')
datarough = soup.find_all('p') 
print(datarough)
```

```python
# So now we can see all instances of text, and our suspicions were confirmed in the reviews/data of interest being
# flanked by <p></p> tags
# Now what?
# Well, we can see that there are only 4 reviews of interest and we know we're going to need to remove the tags
# Why not try using a loop to go from index position 0 in the list that we created above with the find_all() method to position 3?

# Need to initialize a counter variable
i=0

# Also need to initialize an empty list for the loop to fill up as we go through it
data = [0,0,0,0];

# Will now use a while loop to go through our datarough list and get the text from it and place each unique review into 
# a position within our new list

while i<4:

    # You'll notice there are 2 nearly identical lines directly below, with the first, we end up getting a few unwanted 
    # extra characters like \n (new line) formatters , if we don't use strip while  getting the text,
    # these unwanted characters will hang around and may mess up our model when we go to use this data later
    data[i] = datarough[i].get_text()
    
    # As such, it is recommended that to avoid future headaches, use the strip method right as you parse the data to make it easier
    # data[i] = datarough[i].get_text(strip=True)
    data[i] = ''.join(data[i])
    i = i+1

# Now just as a quick sanity check, can see what it is that our data output is, let's look at review 1
print(data[0])
```
